```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
max_node = 384
```

```python
df = pd.read_csv('0_scheduleinfo_list_obj0.csv', sep=',')
```

```python
df.loc[:,'node'] = df.loc[:,'node'] / max_node
```

```python
plt.rcParams["font.size"] = 22 # グラフの文字サイズ

df.plot(drawstyle='steps-post', x='time', y='node', color='red', figsize=(9,7), legend=None)
plt.grid(color='black', linestyle='dashed', linewidth=0.5)
plt.xlabel('Time[h]')
plt.ylabel('Node Usage')
plt.show()
```
