# debug_vis

This script is to visualize the result of randomly generating 1000 jobs. Node size, request time and arrive time interval are generated based on the real data from 'K' Computer.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('debug.csv', delimiter=',')
```

```python
data.ndim
```

```python
data.shape
```

### id

```python
plt.figure(figsize=(7,7), dpi=100) # グラフのサイズ
plt.rcParams["font.size"] = 22 # グラフの文字サイズ

plt.plot(data[:,0])
plt.xlabel('ID')
plt.ylabel('ID')
plt.grid()
plt.show()
```

### index

```python
plt.figure(figsize=(7,7), dpi=100) # グラフのサイズ
plt.rcParams["font.size"] = 22 # グラフの文字サイズ

plt.hist(data[:,1])
plt.xlabel('index')
plt.ylabel('count')
plt.grid(axis='y')
plt.show()
```

### node num [node]

```python
plt.figure(figsize=(7,7), dpi=100) # グラフのサイズ
plt.rcParams["font.size"] = 22 # グラフの文字サイズ

plt.hist(data[:,2])
plt.xlabel('node num')
plt.ylabel('count')
plt.grid(axis='y')
plt.show()
```

### request time [hour]

```python
plt.figure(figsize=(7,7), dpi=100) # グラフのサイズ
plt.rcParams["font.size"] = 22 # グラフの文字サイズ

plt.hist(data[:,3])
plt.xlabel('request time [hour]')
plt.ylabel('count')
plt.grid(axis='y')
plt.show()
```

### arrive time [s]

```python
plt.figure(figsize=(7,7), dpi=100) # グラフのサイズ
plt.rcParams["font.size"] = 22 # グラフの文字サイズ

plt.plot(data[:,4])
plt.xlabel('ID')
plt.ylabel('arrive time [s]')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(7,7), dpi=100) # グラフのサイズ
plt.rcParams["font.size"] = 22 # グラフの文字サイズ

plt.plot(data[:,4]/3600)
plt.xlabel('ID')
plt.ylabel('arrive time [h]')
plt.grid(axis='y')
plt.show()
```

```python
at_interval = np.diff(data[:,4])

plt.figure(figsize=(7,7), dpi=100) # グラフのサイズ
plt.rcParams["font.size"] = 22 # グラフの文字サイズ

plt.plot(at_interval)
plt.xlabel('ID')
plt.ylabel('arrive time interval [s]')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(7,7), dpi=100) # グラフのサイズ
plt.rcParams["font.size"] = 22 # グラフの文字サイズ

plt.hist(at_interval*3600)
plt.grid(axis='y')
plt.xlabel('arrive time interval [s]')
plt.ylabel('count')
plt.show()
```
