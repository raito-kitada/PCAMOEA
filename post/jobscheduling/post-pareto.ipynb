```python
import numpy as np
import matplotlib.colors
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('pareto_objs.csv')
```

```python
obj1, obj2 = data[:, 0], -data[:, 1]
```

```python
fig = plt.figure(figsize=(7,7), dpi=100) # グラフのサイズ
plt.rcParams["font.size"] = 22 # グラフの文字サイズ
sc = plt.scatter(obj1, obj2)
plt.grid()
plt.xlabel('Mean Waiting Time')
plt.ylabel('Mean Node Usage Ratio')
plt.show()
```
