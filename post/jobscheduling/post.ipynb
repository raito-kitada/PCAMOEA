```python
import numpy as np
import matplotlib.colors
import matplotlib.pyplot as plt
```

```python
ngen = 100
```

```python
cmap = plt.get_cmap('rainbow')
```

```python
data = np.loadtxt('0_1_objs.csv')
data = np.insert(data, 2, 1, axis=1)
```

```python
for gen in range(2, ngen+1):
    data2 = np.loadtxt(f'0_{gen}_objs.csv')
    data2 = np.insert(data2, 2, gen, axis=1)
    data = np.vstack((data,data2))
```

```python
data.shape
```

```python
obj1, obj2, gen = data[:, 0], -data[:, 1], data[:, 2]
```

```python
fig = plt.figure(figsize=(7,7), dpi=100) # 
plt.rcParams["font.size"] = 22 # 
sc = plt.scatter(obj1, obj2, c=gen, cmap=cmap, vmin=1, vmax=ngen)
plt.grid()
plt.xlabel('Mean Waiting Time')
plt.ylabel('Mean Node Usage Ratio')
plt.colorbar(sc, label='Gen')
plt.show()
```
