```python
def plot_history(x, y, xlabel, ylabel):
    plt.figure(figsize=(8,7), dpi=100) # figure size
    plt.rcParams["font.size"] = 22 # whole font size

    plt.plot(x, y, marker='o', c='black')
    plt.grid()

    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    plt.show()
```

```python
def scatter2D(x, y, xlabel, ylabel):
    plt.figure(figsize=(8,7), dpi=100) # figure size
    plt.rcParams["font.size"] = 22 # whole font size

    plt.scatter(x, y, marker='o', c='black')
    plt.grid()

    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    plt.show()
```

```python
def scatter2D_history_with_referenceSet(history, referenceSet):
    plt.figure(figsize=(8,7), dpi=100) # figure size
    plt.rcParams["font.size"] = 22 # whole font size

    ites = history.getITESet()

    plt.scatter(x, y, marker='o')
    plt.grid()

    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    plt.show()
```

```python
fig = plt.figure(figsize=(8,7), dpi=100) # figure size
plt.rcParams["font.size"] = 22 # whole font size

ax = fig.add_subplot(1, 1, 1)
ax.set_axisbelow(True)

cmap = plt.get_cmap('rainbow')

attribute_name='GenerationalDistance'
for k, df in meanAccumDatas.items():
    print(k)
    nfes = df['NFE']
    value = df[attribute_name]
    plt.plot(nfes, value, label=k)

plt.grid(ls='--')

plt.xlabel('NFE')
plt.ylabel(attribute_name)
ax.legend(bbox_to_anchor=(1,1.04))

plt.show()
```
