### Function Definitions

#### Utility functions

```python
def makeFileName(header, algorithm, problem, trial, extension):
    if header == '':
        if trial == '':
            return '_'.join([algorithm, problem]) + '.' + extension
        else:
            return '_'.join([algorithm, problem, str(trial)]) + '.' + extension
    else:
        if trial == '':
            return '_'.join([header, algorithm, problem]) + '.' + extension
        else:
            return '_'.join([header, algorithm, problem, str(trial)]) + '.' + extension
```

```python
def readHitoryData(aName, pName, digit, trial):
    fname = '/'.join(
        [base_path, pName, str(digit), makeFileName('history', aName, pName, str(trial), 'txt')]
    )
    df = pd.read_csv(fname, delimiter=' ', header=None, names=colnames)
    return df
```

```python
def readSummaryData(header, aName, pName, digit):
    path = '/'.join(
        [post_path, pName, str(digit)]
    )
    if not os.path.exists(path):
        os.makedirs(path)
    
    fname = os.path.join(path, makeFileName(header, aName, pName, '', 'csv'))
    
    df = pd.read_csv(fname, header=True)
    return df
```

```python
def saveSummaryData(df, header, aName, pName, digit):
    path = '/'.join(
        [post_path, pName, str(digit)]
    )
    if not os.path.exists(path):
        os.makedirs(path)
    
    fname = os.path.join(path, makeFileName(header, aName, pName, '', 'csv'))
    
    df.to_csv(fname, header=True, index=False)
```

```python
def saveFig(header, aName, pName, digit, trial, extension):
    path = '/'.join(
        [post_path, pName, str(digit)]
    )
    if not os.path.exists(path):
        os.makedirs(path)
    
    fname = os.path.join(path, makeFileName(header, aName, pName, str(trial), extension))
    plt.savefig(fname)
```

#### Indicators

$GD(A) = \frac{1}{|A|}\left(\sum_{i=1}^{|A|}d_i^p\right)^{1/p}$, $\quad p=2$

```python
def calcGD_Sphere(solutions):
    gd = 0
    for index, row in solutions.iterrows():
        gd += sum([v*v for v in row]) - 1
    return math.sqrt(gd) / len(solutions)
```

```python
def calcGD_Sphere_Element(solutions):
    gd_elem = np.zeros(len(solutions))
    i = 0
    for index, row in solutions.iterrows():
        gd_elem[i] = sum([v*v for v in row]) - 1
        i += 1
    return gd_elem
```

```python
def calcHV(solutions, ref_point):
    hv = get_performance_indicator("hv", ref_point=ref_point)
    return hv.calc(solutions)
```

#### Main post functions

```python
def calcGDAve_STD(aName, pName, digit):
    # initialize summary
    summary = pd.DataFrame(data=gs, columns=['gen'])
    
    for trial in range(ntrial):
        # read history data in each trial
        df = readHitoryData(aName, pName, digit, trial)
        gen = extractGen(df)
        objs = extractObjs(df, nobj)

        # calc GD 
        GDs = np.zeros(len(gs))
        for i, g in enumerate(gs):
            mask = paretoset(objs[gen<=g], sense=['min'] * nobj)
            reverse_mask = np.logical_not(mask)

            nd_sol = objs[gen<=g][mask]
            GDs[i] = calcGD_Sphere(nd_sol)

        # add GD
        summary[str(trial)] = GDs

    # calc GD average and STD
    GDAve = np.zeros(len(gs))
    GDSTD = np.zeros(len(gs))
    i = 0
    for index, row in summary.iterrows():
        GDAve[i] = np.average(row.iloc[1:(ntrial+1)])
        GDSTD[i] = np.std(row.iloc[1:(ntrial+1)])
        i+=1

    # add results to summary
    summary['Ave'] = GDAve
    summary['STD'] = GDSTD
    
    return summary
```

```python
def post(aName, pName, digit):
    print(aName+'_'+pName+'_'+str(digit))
    
    # calc GD average and std
    summary = calcGDAve_STD(aName, pName, digit)

    # save summary
    saveSummaryData(summary, 'GD', aName, pName, digit)
    
    return summary
```
