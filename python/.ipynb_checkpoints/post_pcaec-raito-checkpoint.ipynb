```python
# 標準ライブラリ
import os

# サードパーティ製ライブラリ
import matplotlib.pyplot as plt
import paretoset

import pandas as pd

# その他のライブラリ
import lab.moea.post as lmp
import lab.moea.viz as lmv

# 作業ディレクトリの確認
os.getcwd()
```

```python
# 計算結果の起点となるパス
output_base_path = '../output/pcaec/'

# リファレンスデータセット（パレートフロント・データセット）のベースパス
refset_base_path = '../pf'

# 目的関数の数
nobj = 2
# ベンチマーク問題
pName = 'DTLZ2'
# 最適化アルゴリズム
aNames = ['NSGAII-LAB', 'NSGAIIPOD']
aNamesLabel = {'NSGAII-LAB':'NSGA-II', 'NSGAIIPOD':'with PCA'}
# 試行回数
ntrial = 10

# リファレンスデータセットのファイル名を含むパス（DTLZ, WFGシリーズの場合）
refset_path = '{}/{}.{}D.pf'.format(refset_base_path, pName, str(nobj))

# 計算結果のパス（目的関数を追加。この下にアルゴリズム/試行
base_path = '{}/{}/'.format(output_base_path, pName)
```

```python
# 評価値の平均値を計算しファイルに保存する(1度だけ実行すればよい)
lmp.makeMeanAccumDatas(base_path, aNames, ntrial)
```

```python
# 計算した平均値の読み込み
meanAccumDatas = lmp.loadMeanAccumDatas(base_path, aNames)
```

```python
# 評価値の平均値の可視化
fig = lmv.plot_accumDatas(meanAccumDatas, aNamesLabel, 'NFE', 'GenerationalDistance', 'NFE', 'GD', 'upper left', (1, 1))
print(type(fig))
fig = lmv.plot_accumDatas(meanAccumDatas, aNamesLabel, 'NFE', 'InvertedGenerationalDistance', 'NFE', 'IGD', 'upper left', (1, 1))
fig = lmv.plot_accumDatas(meanAccumDatas, aNamesLabel, 'NFE', 'Hypervolume', 'NFE', 'HV', 'upper left', (1, 1))
```

```python
# 評価値の平均値の可視化
fig = lmv.plot_accumDatas(meanAccumDatas, 'ITE', 'GenerationalDistance', 'ITE', 'GD', 'upper left', (1, 1))
fig = lmv.plot_accumDatas(meanAccumDatas, 'ITE', 'InvertedGenerationalDistance', 'ITE', 'IGD', 'upper left', (1, 1))
fig = lmv.plot_accumDatas(meanAccumDatas, 'ITE', 'Hypervolume', 'ITE', 'HV', 'upper left', (1, 1))
```
