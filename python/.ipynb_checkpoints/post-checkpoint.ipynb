```python
import lab.moea.post as lmp
import lab.moea.viz as lmv
import matplotlib.pyplot as plt
```

```python
output_base_path = '../output/basic_sample/'

pName = 'DTLZ3'
# aNames = ['CMAES', 'eMOEA', 'eNSGAII', 'GDE3', 'IBEA', 'MOEAD',
# 'NSGAII', 'NSGAIII', 'OMOPSO', 'PAES', 'PESA2', 'Random', 'SPEA2']
aNames = ['MOEAD', 'NSGAII', 'NSGAIII', 'Random']

base_path = '{}/{}'.format(output_base_path, pName)
```

```python
meanAccumDatas = lmp.loadMeanAccumDatas(base_path, aNames)
```

```python
def plot_accumDatas(accumDatas, attribute_name, ylabel, 
                    loc='upper left', anchor=(1, 1)):
    """Plot accumulator dataset
    
    """

    fig = plt.figure(figsize=(8, 7), dpi=100)  # figure size
    plt.rcParams["font.size"] = 22  # whole font size
    # plt.subplots_adjust(left=0.14, top=0.96, right=0.95, bottom=0.12)

    ax = fig.add_subplot(1, 1, 1)
    ax.set_axisbelow(True)

    # cmap = plt.get_cmap('rainbow')

    for k, df in accumDatas.items():
        nfes = df['NFE']
        value = df[attribute_name]
        plt.plot(nfes, value, label=k)

    plt.grid(ls='--')

    plt.xlabel('NFE')
    plt.ylabel(ylabel)
    ax.legend(loc=loc, bbox_to_anchor=anchor)

    plt.show()

    return fig
```

```python
fig = plot_accumDatas(meanAccumDatas, 'GenerationalDistance', 'GD', 'upper left', (1, 1))
# fig.savefig('result.pdf', bbox_inches = "tight")
```

```python
lmv.plot(meanAccumDatas['NSGAII']['NFE'], meanAccumDatas['NSGAII']['GenerationalDistance'], 'NFE', 'GD')
```

```python
def plot(x, y, xlabel, ylabel):
    """Draw simple line plot

    Args:
        x (array-like): horizontal coordinates
        y (array-like): virtical coordinates
        xlabel (str): label of horizontal axis
        ylabel (str): label of virtical axis

    Returns:
        Figure object
    """

    fig = plt.figure(figsize=(8, 7), dpi=100)  # figure size
    plt.rcParams["font.size"] = 22  # whole font size
    # plt.subplots_adjust(left=0.14, top=0.96, right=0.95, bottom=0.12)

    ax = fig.add_subplot(1, 1, 1)
    ax.set_axisbelow(True)

    plt.plot(x, y, marker='o', c='black')
    plt.grid(ls='--')

    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    plt.show()

    return fig
```

```python
fig = plot(meanAccumDatas['NSGAII']['NFE'], meanAccumDatas['NSGAII']['GenerationalDistance'], 'NFE', 'GD')
```
